## Data Sources and Integration

- The **Purchases** table records individual purchase transactions, including purchase date, product (brand) details, vendor information, quantities purchased, and amounts paid.  
- The **Purchase_Prices** table provides product-level pricing details, including both actual and purchase prices, with each vendorâ€“brand combination stored uniquely.  
- The **Vendor_Invoice** table consolidates data from *Purchases*, summarizing purchase quantities and amounts, while also including freight costs. Each record is uniquely identified by vendor and PO number.  
- The **Sales** table captures sales transactions, detailing the brands purchased from vendors, quantities sold, selling prices, and revenue generated.  

Since the required data is distributed across multiple tables, a **summary table** is created to support the analysis. This integrated table combines:  
- Vendor purchase transactions  
- Sales transaction data  
- Freight costs per vendor  
- Actual product prices


```python
vendor_sales_summary = pd.read_sql_query("""WITH FreightSummary AS (
    SELECT
        VendorNumber,
        SUM(Freight) AS FreightCost
    FROM vendor_invoice
    GROUP BY VendorNumber
),

PurchaseSummary AS (
    SELECT 
        p.VendorNumber,
        p.VendorName,
        p.Brand,
        p.Description,
        p.PurchasePrice,
        pp.Price AS ActualPrice,
        pp.Volume, 
        SUM(p.Quantity) AS TotalPurchaseQuantity,
        SUM(p.Dollars) AS TotalPurchaseDollars
    FROM purchases p
    JOIN purchase_prices pp
        ON p.Brand = pp.Brand
    WHERE p.PurchasePrice > 0
    GROUP BY p.VendorNumber, p.VendorName, p.Brand, p.Description, p.PurchasePrice, pp.Price, pp.Volume
),

SalesSummary AS (
    SELECT
        VendorNo,
        Brand, 
        SUM(SalesQuantity) AS TotalSalesQuantity,
        SUM(SalesDollars) AS TotalSalesDollars,
        SUM(SalesPrice) AS TotalSalesPrice,
        SUM(ExciseTax) AS TotalExciseTax
    FROM sales
    GROUP BY VendorNo, Brand
)

SELECT
    ps.VendorNumber,
    ps.VendorName, 
    ps.Brand, 
    ps.Description, 
    ps.PurchasePrice,
    ps.ActualPrice, 
    ps.Volume,
    ps.TotalPurchaseQuantity,
    ps.TotalPurchaseDollars,
    ss.TotalSalesQuantity,
    ss.TotalSalesDollars,
    ss.TotalSalesPrice,
    ss.TotalExciseTax,
    fs.FreightCost
FROM PurchaseSummary ps
LEFT JOIN SalesSummary ss
    ON ps.VendorNumber = ss.VendorNo
    AND ps.Brand = ss.Brand
LEFT JOIN FreightSummary fs
    ON ps.VendorNumber = fs.VendorNumber
ORDER BY ps.TotalPurchaseDollars DESC""",conn)
```

## Data Preprocessing

The dataset was prepared for analysis through the following steps:

1. **Checking for Null Values:** Identified missing or incomplete entries across all tables to ensure data quality.  
2. **Data Type Verification:** Ensured that each column had the correct data type (e.g., numeric, date, text) for accurate computations and analysis.  
3. **Filling Missing Values:** Replaced missing numeric values with `0` where appropriate to maintain consistency in calculations.  
4. **Trimming Text Values:** Removed leading and trailing spaces in text fields to standardize categorical data.  
5. **Adding Calculated Columns:** Created new columns such as **Gross Profit**, **Profit Margin**, **Sales to purchase ratio**, and **Stock Turnover** to derive key performance metrics for analysis.

## Adding calculated columns

```python
vendor_sales_summary['GrossProfit'] = vendor_sales_summary['TotalSalesDollars'] - vendor_sales_summary['TotalPurchaseDollars']
```

```python
vendor_sales_summary['ProfitMargin'] = (vendor_sales_summary['GrossProfit'] / vendor_sales_summary['TotalSalesDollars'])*100
```

```python
vendor_sales_summary['StockTurnover'] = vendor_sales_summary['TotalSalesQuantity']/vendor_sales_summary['TotalPurchaseQuantity']
```

```python
vendor_sales_summary['SalestoPurchaseRatio'] = vendor_sales_summary['TotalSalesDollars']/vendor_sales_summary['TotalPurchaseDollars']
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("""CREATE TABLE vendor_sales_summary (
    VendorNumber INT,
    VendorName VARCHAR(100),
    Brand INT,
    Description VARCHAR(100),
    PurchasePrice DECIMAL(10,2),
    ActualPrice DECIMAL(10,2),
    Volume,
    TotalPurchaseQuantity INT,
    TotalPurchaseDollars DECIMAL(15,2),
    TotalSalesQuantity INT,
    TotalSalesDollars DECIMAL(15,2),
    TotalSalesPrice DECIMAL(15,2),
    TotalExciseTax DECIMAL(15,2),
    FreightCost DECIMAL(15,2),
    GrossProfit DECIMAL(15,2),
    ProfitMargin DECIMAL(15,2),
    StockTurnover DECIMAL(15,2),
    SalesToPurchaseRatio DECIMAL(15,2),
    PRIMARY KEY (VendorNumber, Brand)
);
""") 
```

```python
vendor_sales_summary.to_sql('vendor_sales_summary', conn, if_exists = 'replace', index= False)
```
